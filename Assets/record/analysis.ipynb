```python
import numpy as np
import matplotlib.pyplot as plt

x_path = './x_err.txt' 
x_data = np.loadtxt(x_path) 
y_path = './y_err.txt' 
y_data = np.loadtxt(y_path) 
length = x_data.size - 1500
```

```python
t = np.linspace(0, length/200, length)

plt.plot(t, x_data[0:length])
plt.plot(t, np.zeros(length))
plt.xlabel('Time (s)')
plt.ylabel('Horizontal Error (m)')
plt.show()
```

```python
plt.plot(t, y_data[0:length])
plt.plot(t, np.zeros(length))
plt.xlabel('Time (s)')
plt.ylabel('Vertical Error (m)')
plt.show()
```
